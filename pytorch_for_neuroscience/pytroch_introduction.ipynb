# PyTorch for Neuroscience
An open source machine learning framework that allows you to write your own neural networks and optimize them efficiently.

Why write neural networks in Neuroscience?  
Three basic reasons
- As a computational tool to analyse brain data
- Model for the brain
- Bridging biology and artificial intelligence

**What other options exists and why PyTorch?**  
Other options such as MATLAB, NEST, R, Julia etc., have been used for specific purposes. We use PyTorch for its flexibility, dynamic computation graph and extensive ecosystem for modeling, analysing and understanding neural data.  
PyTorch is not the only framework of its kind, alternatives include TensorFlow, JAX and Caffe. There are many tutorials and resources that provide a great introduction to PyTorch. Here, we explore PyTorch use in various Neuroscience concepts.

## PyTorch basics
- **Tensor Library**  
Data type at the core of PyTorch are Tensors. Tensors are mathematical objects that can be characterized by their order (or rank), which provides the number of dimensions. Tensors are the equivalent to Numpy arrays. It has an additional feature of accelerated computation on GPUs
- **Automatic Differentiation**  
Calculating derivatives is the crucial step in all the optimization algorithms that we need to train deep networks. To calculate derivatives, automatic differentiation works backwards through a computational graph by applying the chain rule. We will use PyTorch to automatically get the gradients.  
- **Deep Learning Library**  
Deep learning models have several elements that work together to produce a trained mode: model, optimizer, loss, trainer. PyTorch's deep learning library utilities makes use of the tensor library and automatic differentiation to offer modular, flexible and efficient builduing blocks.
